```python
#importing in blight violation data
blight_violations = pd.read_csv('/content/drive/MyDrive/detroit data/blight_violations (1).csv')
blight_violations.head(1)
#parcelno
```

```python
#now we have to convert the unix timestamp in the violation_date to a readbable date (using ms) becuase its 13 digits i think
blight_violations_1['violation_date'] = pd.to_datetime(blight_violations_1['violation_date'], unit='ms')

blight_violations_1.head()
```

### POLICE DATA

```python
police =pd.read_csv('/content/drive/MyDrive/detroit data/Police_Serviced_911_Calls_805604165310935371.csv')
police1 = police.loc[:,['Code Description','Category','Priority','Call Time','Neighborhood Name','Council District']]
```

```python
group1 = ['P', '1','2','3']
def p1(Priority):
  if Priority in group1:
    return 1
  else:
    return 0
police1['High_Priority'] = police1['Priority'].apply(p1)
police1.head()
```

```python
police1[['Month','Day', 'temp']] = police1['Call Time'].str.split('/',expand=True)
police1=police1.drop(columns=['Call Time','Month','Day'])
def getYear(y):
 if y:
  return y[0:4]

police1['Year'] = police1['temp'].apply(getYear)
police1.head()
```

```python
police1 = police1.loc[police1['Year'].isin(['2019', '2020', '2021', '2022', '2023'])]
policd1 = police1.loc[police1['High_Priority'] == 1]
police1 = police1.drop(columns=['temp','Code Description','Category','Priority','Council District'])
police1.head()
```

```python
count_df = police1.groupby(['Year', 'Neighborhood Name']).size().reset_index(name='count')
finalpolice_df = count_df.groupby('Neighborhood Name')['count'].mean().reset_index(name='yearly_average')
#gives yearly high priority serviced calls by neighborhood
finalpolice_df.head()
```
