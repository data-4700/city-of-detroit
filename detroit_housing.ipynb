### POLICE DATA

```python
police =pd.read_csv('/content/drive/MyDrive/detroit data/Police_Serviced_911_Calls_805604165310935371.csv')
police1 = police.loc[:,['Code Description','Category','Priority','Call Time','Neighborhood Name','Council District']]
```

```python
group1 = ['P', '1','2','3']
def p1(Priority):
  if Priority in group1:
    return 1
  else:
    return 0
police1['High_Priority'] = police1['Priority'].apply(p1)
police1.head()
```

```python
police1[['Month','Day', 'temp']] = police1['Call Time'].str.split('/',expand=True)
police1=police1.drop(columns=['Call Time','Month','Day'])
def getYear(y):
 if y:
  return y[0:4]

police1['Year'] = police1['temp'].apply(getYear)
police1.head()
```

```python
police1 = police1.loc[police1['Year'].isin(['2019', '2020', '2021', '2022', '2023'])]
policd1 = police1.loc[police1['High_Priority'] == 1]
police1 = police1.drop(columns=['temp','Code Description','Category','Priority','Council District'])
police1.head()
```

```python
count_df = police1.groupby(['Year', 'Neighborhood Name']).size().reset_index(name='count')
finalpolice_df = count_df.groupby('Neighborhood Name')['count'].mean().reset_index(name='yearly_average')
#gives yearly high priority serviced calls by neighborhood
finalpolice_df.head()
```

```python
#replacing NA values in fine_amount to make life easy
df = clean
df['fine_amount'] = df['fine_amount'].fillna(0)
df.head(5)
```

```python
#gathering number of houses by neighborhood
housecount = df.groupby('Neighborhood').size().reset_index(name='count')
hcount = housecount.groupby('Neighborhood')['count'].mean().reset_index(name='neighborhood_population')
hcount.head(5)
```

```python
#gathering number of vacant houses by neighborhood
num_vacant_house = df.groupby('Neighborhood')['vacant'].apply(lambda x: (x==1).sum()).reset_index(name='num_vacant_neighborhood')
num_vacant_house.head()
```

```python
#merging cleaned dataset on neighborhood population info
merged1 = pd.merge(df, hcount, how='left', on='Neighborhood')
merged1.head()
```

```python
#merging number of vacant houses in neighborhood
merged2 =  pd.merge(merged1, num_vacant_house, how='left', on='Neighborhood')
merged2.head()
```

```python
#normalizing crime by neighborhood population, removing outliers normalized crime outliers (lost roughly 100 nonvacant homes, 3 vacant homes from this removal)
final = merged2
final['normcrime'] = final['yearly_average']/final['neighborhood_population']
temp = merged2[['Neighborhood','normcrime','yearly_average','neighborhood_population']]
#Lafeyette Park, Rivertown, Brush Park, #Forest Park, Jeffries, Milwaukee Junction, Palmer Park, Rouge Park, West Side Industrial
df_final = final.loc[~try2['Neighborhood'].isin(['Lafayette Park', 'Rivertown','Brush Park', 'Forest Park', 'Jeffries', 'Milwaukee Junction', 'Palmer Park', 'Rouge Park', 'West Side Industrial'])]
df_final.head()
```
